## Menentukan Pertanyaan Bisnis

**SMART Question** adalah sebuah framework untuk merumuskan pertanyaan secara terstruktur agar memeperoleh informasi yang mendalam.

* **Spesific (Spesifik)**
  * Pertanyaan harus jelas, fokus pada sebuah topik tertentu, dan tidak bermakna ganda. Hindari pertanyaan yang terlalu luas.
    * Salah: Bagaimana cara meningkatkan penjualan?
    * Benar: Faktor apa saja yang memengaruhi penurunan penjualan produk kategori elektronik di wilayah Jakarta selama kuartal terakhir?
* **Measurable (Terukur)**
  * Pertanyaan harus bisa dijawab dengan angka atau matrix yang konkret, harus tahu apa yang akan dihitung.
    * Salah: Apakah pelanggan senang dengan layanan kita?
    * Benar: Berapa skor rata-rata Customer Satisfaction untuk layanan purna jual bulan ini dibandingkan bulan lalu?
* **Action-Oriented (Berorientasi Aksi)**
  * Hasil dari pertanyaan harus bisa memberikan arahan untuk melakukan tindakan nyata. Jika pertanyaan terjawab, stakeholder harus tahu apa langkah selanjutnya.
    * Salah: Mengapa orang suka berbelanja?
    * Benar: Fitur apa pada aplikasi yang paling sering digunakan sebelum pengguna memutuskan untuk melakukan checkout?
* **Relevant (Relevan)**
  * Hasil dari pertanyaan harus sejalan dengan tujuan utama bisnis atau masalah yang sedang dihadapi.
    * Salah: Menanyakan tentang stok gudang saat masalah utamanya adalah efektivitas kampanye media sosial.
    * Benar: Apakah kampanye iklan di Instagram memberikan Return on Ad Spend (ROAS) yang lebih tinggi dibandingkan iklan di TikTok?
* **Time-bound (Terikat Waktu)**
  * Pertanyaan harus ada batasan waktu yang jelas agar analisis memiliki konteks yang tepat.
    * Salah: Berapa banyak pengguna baru kita?
    * Benar: Berapa tingkat pertumbuhan pengguna baru secara bulanan (Month-over-Month) sepanjang tahun 2025?

**Contoh pertanyaan bisnis yang memenuhi seluruh elemen SMART**

Keterangan:

- **Specific**: Fokus pada "penurunan conversion rate" untuk "aplikasi Android" di "Jabodetabek". Bukan sekadar penjualan turun.
- **Measurable**: Ada angka konkret yang ingin dianalisis, yaitu penurunan sebesar "5%".
- **Action-Oriented**: Dengan mengetahui faktor penyebabnya misalnya bug pada versi Android tertentu atau kendala logistik di Jabodetabek, tim bisa langsung melakukan perbaikan teknis atau operasional.
- **Relevant**: Penurunan konversi saat Flash Sale adalah masalah kritis bagi bisnis retail/e-commerce.
- **Time-bound**: Dibatasi pada periode spesifik "Flash Sale Maret 2026".

## Visualization & Explanatory Analysis

### Pertanyaan 1: Faktor apa saja yang memengaruhi penurunan jumlah peminjaman sepeda (cnt) pada musim dingin dan saat kondisi cuaca buruk selama periode tahun 2011–2012?

```python
#Rata-rata peminjaman sepeda berdasarkan season
season_rentals = day_df.groupby("season")["total_rentals"].mean().reset_index()

plt.figure(figsize=(10, 6))
sns.barplot(data=season_rentals, x="season", y="total_rentals")
plt.title("Average Bike Sharing by Season")
plt.xlabel("Season")
plt.ylabel("Total Rentals")
plt.show()

#Rata-rata peminjaman sepeda berdasarkan weather
weather_rentals = day_df.groupby("weathersit")["total_rentals"].mean().reset_index()

plt.figure(figsize=(10, 6))
sns.barplot(data=weather_rentals, x="weathersit", y="total_rentals")
plt.title("Average Bike Sharing by Weather")
plt.xlabel("Weather")
plt.ylabel("Total Rentals")
plt.show()

#Map korelasi
plt.figure(figsize=(10, 6))
sns.heatmap(day_df.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

**Insight:** (Opsional)
- Berdasarkan hasil analisis, penurunan jumlah peminjaman sepeda dipengaruhi oleh musim, cuaca, serta faktor lingkungan seperti suhu, kelembapan, dan kecepatan angin. Musim dingin menunjukkan jumlah peminjaman yang lebih sedikit dibandingkan dengan musim lainnya, sementara cuaca cerah (weathersit = 1) memiliki jumlah peminjaman tertinggi dan cuaca buruk (weathersit =3) memiliki jumlah peminjaman terendah. Hasil heatmap juga menunjukkan bahwa suhu yaitu variabel temp dan atemp memiliki korelasi positif terhadap jumlah peminjaman, sedangkan weathersit, windspeed, dan hum memiliki korelasi negatif. Dengan demikian, ditunjukkan bahwa kondisi cuaca yang dingin, lembap, dan berangin membuat kurang nyaman sehingga menurunkan minat masyarakat dalam menggunakan layanan.

## Analisis Lanjutan (Opsional): Clustering

```python
#menghitung rata-rata peminjaman per jam
hourly_rentals = hour_df.groupby("hour")["total_rentals"].mean().reset_index()
hourly_rentals
```

```python
#Membuat fungsi kategori
def demand_category(total_rentals):
  if total_rentals >= 300:
    return "High Demand"
  elif total_rentals >= 100:
    return "Medium Demand"
  else:
    return "Low Demand"
```

```python
#Menambah kolom cluster
hourly_rentals["demand_cluster"] = hourly_rentals["total_rentals"].apply(demand_category)
hourly_rentals
```

```python
#Visualisasi
plt.figure(figsize=(10, 5))
sns.barplot(data=hourly_rentals, x="hour", y="total_rentals", hue="demand_cluster")

plt.title("Bike Sharing Demand Cluster by Hour")
plt.xlabel("Hour")
plt.ylabel("Average Total Rentals")
plt.show()
```

Selain itu, perusahaan juga bisa memanfaatkan data prakiraan cuaca untuk membantu perencanaan distribusi sepeda secara lebih efisien.
- Perusahaan dapat meningkatkan jumlah unit sepeda dan kapasitas layanan pada pertengahan tahun dan musim gugur untuk mengurangi kekurangan unit. Pada periode permintaan rendah, perusahaan dapat menjalankan program promosi atau diskon untuk menjaga tingkat penggunaan layanan tetap stabil.
- Perusahaan perlu memprioritaskan ketersediaan sepeda pada jam sibuk pagi dan sore hari dengan melakukan redistribusi unit dengat terencana. Selain itu, pengurangan operasional pada jam dengan permintaan sanga rendah dapat membantu meningkatkan efisiensi biaya operasional dan pemeliharaan layanan.
